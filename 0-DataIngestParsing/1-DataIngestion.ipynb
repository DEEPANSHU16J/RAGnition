#### TEXTLOADER - read single file 

```python
from langchain_community.document_loaders import TextLoader

Loader = TextLoader(r"data\txt_files\Antino_Intro.txt")
documents = Loader.load()

documents
print(type(documents))
print(documents)
```

#### DIRECTORY LOADER - read multiple files

```python
from langchain_community.document_loaders import DirectoryLoader

#load all files from directory

dir_loader = DirectoryLoader(r"data\txt_files",
                             glob="**/*.txt",  #patter to match all the files
                             loader_cls = TextLoader, #Loader class used for loading files 
                             loader_kwargs={"encoding":"utf-8"},
                             show_progress=True 
                             )  

documents  = dir_loader.load()

documents
print(type(documents))
print(documents)
```

#### TEXT SPLITTING STRATERGIES IN LANGCHAIN

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)

print(documents)
```

##### Method 1- character text splitter 

```python

text = documents[0].page_content 
text
```

```python
#METHOD 1 character text splitter 
print("CHARACTER BASED SPLITTING")
char_splitting = CharacterTextSplitter(
    separator='\n',
    chunk_size = 200,
    chunk_overlap = 20,
    length_function = len
)

char_chunks  = char_splitting.split_text(text)
print(f"created {len(char_chunks) } chunks")

print(f"chunk 0 : length:{len(char_chunks[0])} : {char_chunks[0]}")
print(f"chunk 1 : length:{len(char_chunks[1])} : {char_chunks[1]}")
```

##### Method 2 - recursive character text splitter        {Better Approach}


```python
print("Recursive character based splitting")
recursive_splitting  = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""],
    chunk_size = 200, 
    chunk_overlap = 20,
    length_function = len
)

recursive_chunks = recursive_splitting.split_text(text)
print(f"created {len(recursive_chunks)} chunks ")
print(f"chunk 0 : length:{len(recursive_chunks[0])} : {recursive_chunks[0]}")
print(f"chunk 1 : length:{len(recursive_chunks[1])} : {recursive_chunks[1]}")
```

##### Method 3 - Token based splitting 

```python
print("Token based splitting ")
token_splitter = TokenTextSplitter(
    chunk_size = 50,
    chunk_overlap  =10
)


token_chunks  = token_splitter.split_text(text)
print(f"created {len(token_chunks)} chunks ")
print(f"chunk 0 : length : {len(token_chunks[0])} : {token_chunks[0]}")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print(f"chunk 1 : length : {len(token_chunks[1])} : {token_chunks[1]}")
```
